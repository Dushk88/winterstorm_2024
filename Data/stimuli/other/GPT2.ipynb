```python

!pip install transformers
!pip install progressbar2


import torch
import string
import progressbar
import matplotlib.pyplot as plt
import numpy as np
from transformers import GPT2Tokenizer, TFGPT2LMHeadModel
import tensorflow as tf

tokenizer = GPT2Tokenizer.from_pretrained("gpt2-large")
model = TFGPT2LMHeadModel.from_pretrained("gpt2-large")

BOS_TOKEN = "<|endoftext|>"
EOS_TOKEN = "<|endoftext|>"
```

```python

story1="if you happened to find yourself on the banks of the Ohio river on a particular afternoon in the spring of eighteen o six somewhere just to the north of Wheeling West Virginia say you would probably have noticed a strange makeshift craft drifting lazily down the river at the time this particular stretch of the Ohio wide and brown and bounded on both sides by steep shoulders of land thick with oaks and hickories fairly boiled with river traffic as a ramshackle armada of keelboats and barges ferried settlers from the comparative civilization of Pennsylvania to the wilderness of the Northwest Territory the peculiar craft you would have caught sight of that afternoon consisted of a pair of hollowed out logs that had been lashed together to form a rough catamaran a sort of canoe plus sidecar in one of the dugouts lounged the figure of a skinny man of about thirty who may or may not have been wearing a burlap coffee sack for a shirt and a tin pot for a hat according to the man in Jefferson county who deemed the scene worth recording the fellow in the canoe appeared to be snoozing without a care in the world evidently trusting in the river to take him wherever it was he wanted to go"

story1 = BOS_TOKEN+story1+EOS_TOKEN
inputs_story = tokenizer(story1, return_tensors="tf")
inputs_story_token = tokenizer.tokenize(story1)


outputs = model(inputs_story)
logits = outputs.logits
predictions = tf.math.softmax(logits, axis=-1)

ids = tokenizer.encode(story1, return_tensors="tf")
#textfile = open('/content/gdrive/My Drive/Colab Notebooks/GPT2/'+story_set[ss]+ '_GPT2.txt', "w")

for i in range(1,len(predictions[0,:,1])):
s=predictions[0,i-1,ids[0,i]]
print([inputs_story_token[i], -np.log2(s.numpy())])
#textfile.write(inputs_story_token[i]+"\t"+str(-np.log2(s.numpy())) + "\n")

#textfile.close()
```
